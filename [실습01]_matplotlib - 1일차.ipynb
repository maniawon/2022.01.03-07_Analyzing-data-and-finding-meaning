## 9) 추가 ③ 수평선, 수직선 추가하기  
- 수평선 추가  
plt.axhline(y값, color, linestyle)  
- 수직선 추가  
plt.axvline(x값, color, linestyle)

① 다음의 조건에 따라 라인차트를 그리기 위한 코드를 작성하고, 수평선, 수직선을 추가해 봅시다.  
- 데이터 
   - 'v1': [1, 2, 3, 4, 5]  
   - 'v2': [2, 5, 3, 1, 2]  
- v1, v2의 값으로 지정  
- 레이블: mango  
- 수평선  
   - y값: 3.3  
   - color: grey  
   - line style: dashed  
- 수직선  
   - x값: 2.4

   - line style: dashed

```python
dict1 = {'v1':[1, 2, 3, 4, 5],
         'v2':[2, 5, 3, 1, 2]}
print(dict1)

plt.plot('v1', 'v2', 'ro--', data = dict1, label = 'mango')

plt.legend()

# 수평선 추가
plt.axhline(3.3, color = 'grey', linestyle = 'dashed')

# 수직선 추가
plt.axvline(2.4, color = 'grey', linestyle = 'dashed')
plt.savefig('a.png')
```

## 11) 추가 ⑤ 여러 그래프 나눠서 그리기  
plt.subplot(rows, cols, index)  
index는 1부터 시작한다. 0쓰면 error 뜬다.      

plt.tight_layout() ← 여러 그래프 그릴 때 보기 좋게!

① 다음의 그래프들을 위아래로 나누어 그리시오.  
- 데이터  
   - 'v1': [1, 2, 3, 4, 5]  
   - 'v2': [2, 5, 3, 1, 2]  
   - 'v3': [4, 6, 7, 6, 5]  
- 그래프 1  
    - v1, v2의 값으로 지정  
- 그래프 2  
    - v1, v3의 값으로 지정

```python
dict1 = {'v1': [1, 2, 3, 4, 5], 'v2': [2, 5, 3, 1, 2],
         'v3': [4, 6, 7, 6, 5]}
print(dict1)

plt.subplot(2, 1, 1) # 1행 1열
plt.plot('v1', 'v2', 'go--', data = dict1)

plt.subplot(2, 1, 2) # 2행 1열
plt.plot('v1', 'v3', 'rs-', data = dict1)

# plt.tight_layout() ← 여러 그래프 그릴 때 보기 좋게
plt.tight_layout()

plt.savefig('a.png')
```
