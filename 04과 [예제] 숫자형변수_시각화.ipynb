* [4-1] 히스토그램 : plt.hist(1차원데이터 , bins = 구간의 개수, edgecolor = ‘gray’)  
→ 숫자형 변수의 분포를 살펴보기 위한 가장 기본이 되는 그래프

```python
# 0) 라이브러리 로딩

import matplotlib.pyplot as plt

# 1) 히스토그램 기본

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
plt.hist(age, bins = 5, edgecolor = 'gray')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2) 구간의 갯수를 조정해 봅시다. ← bins = 지정할 bins수

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
plt.hist(age, bins = 10, edgecolor = 'gray')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# 3) 구간 값과 갯수를 확인해 봅시다.

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
hist1 = plt.hist(age, bins = 5, edgecolor = 'gray')

print('plt.hist를 저장하면 튜플: {}'.format(type(hist1)))
print('plt.hist를 저장하면 튜플 안에는 Numpy array가 2개로 구성:\n {}'.format(hist1))

print('=' * 50)
print('빈도수 : ', hist1[0]) # 첫번째 Numpy array : 빈도수
print('구간값 : ', hist1[1]) # 두번째 Numpy array : 구간값
```

```python
# 4) 연습문제를 풀어 봅시다.
import pandas as pd
data = pd.read_csv("https://bit.ly/3HaMAtZ", usecols = ['Age','Fare']) #  usecols: 가져올 컬럼 지정

# Age에 칼럼에 대해 히스토그램을 그려 봅시다.
import matplotlib.pyplot as plt
# bins = 8, 16, 32, 64
plt.hist(data['Age'], bins = 64, edgecolor = 'gray')
plt.show()
```

```python
# Age에 칼럼에 대해 히스토그램을 그릴때, 
# 빈도수와 구간값을 살펴 봅시다.
# bins = 8
result = plt.hist(data['Age'], bins = 8, edgecolor = 'gray')
print(result[0]) # 빈도수
print(result[1]) # 구간값
```

```python
# Age 불러오는 방법
print("data['Age']로 불러오기:\n {}".format(data['Age'])) # 권장사항
print("data.Age 로 불러오기:\n {}".format(data.Age)) # 이렇게 불러와도 되긴된다
```

* [4-2] 밀도함수(Kernel Density Estimation) 그래프 : plt.plot(kind = 'kde')

```python
# 0) 라이브러리 로딩

import matplotlib.pyplot as plt
import pandas as pd

# ----------------------------------------------
# 1) 밀도함수 그래프.

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
age = pd.Series(age)
age.plot(kind = 'kde')
plt.show()
```

```python
# 2) 연습문제

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]

# 히스토그램을 그리시오.
plt.hist(age, bins = 10, edgecolor = 'gray')
plt.show()

# 밀도함수 그래프를 그리시오.
age1 = pd.Series(age)
age1.plot(kind = 'kde') # pandas를 이용한 차트 그리기
plt.show()
```

- 두 그래프는 어떤 차이가 있나요?  
→ 히스토그램은 구간으로 보이고  
→ 밀도함수 그래프는 연속으로 보인다

```python
# 추가
data = pd.read_csv("https://bit.ly/3HaMAtZ", usecols = ['Age','Fare']) #  usecols: 가져올 컬럼 지정
print(data.head())

# 기초 통계랑
print('기초 통계랑:\n', data['Fare'].describe())

# histogram
plt.hist(data['Fare'], bins = 50, edgecolor = 'gray')
plt.show()

# density plot
data['Fare'].plot(kind = 'kde')
plt.show()
```

* [4-3] boxplot : plt.boxplot(data, vert = True) 
- boxplot을 저장하면 딕셔너리로 나온다.
    - boxplot 세로 : vert = True &  각 key에 해당되는 정보를 얻기 위해서는 .get_ydata() 메서드
    - boxplot 가로 : vert = False & 각 key에 해당되는 정보를 얻기 위해서는 .get_xdata() 메서드

```python
# 1) boxplot 기본 ← 디폴트값 : vert = True

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
plt.boxplot(age, vert = True)
plt.show()
```

```python
# 2) 옆으로 그리기

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
plt.boxplot(age, vert = False)
plt.show()
```

```python
box = plt.boxplot(age, vert = False)
box['whiskers'][0].get_xdata()[1] # 아래쪽 수염의 min
```

→ min과 (1사분위수 - 1.5 * IQR) 같다

```python
# 3) 박스플로의 값들 가져오기

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
box1 = plt.boxplot(age)

print('plt.boxplot를 저장하면 튜플: {}'.format(type(box1)))
print('plt.boxplot를 저장하면 각 key에 해당 되는 정보를 얻을 수 있다:\n'.format(box1.keys()))

print(box1['whiskers'])
print('아래쪽 수염의 max, min :', box1['whiskers'][0].get_ydata()) # 아래쪽 수염의 max, min
print('위쪽 수염의 min, max', box1['whiskers'][1].get_ydata()) # 위쪽 수염의 min, max
```

```python
# 4) 연습문제를 풀어 봅시다.
import pandas as pd
data = pd.read_csv("https://bit.ly/3HaMAtZ", usecols = ['Age','Fare'])

# Age에 칼럼에 대해 
print(data['Age'].describe())

# 박스 플롯을 그려 봅시다.
plt.boxplot(data['Age'], vert = True)
plt.show()

# 박스 플롯을 옆으로 그려 봅시다.
plt.boxplot(data['Age'], vert = False)
plt.show()

# 박스 플롯의 아래 수염의 min, max, 윗 수염의 min, max
box = plt.boxplot(data['Age'])
print(box['whiskers'][0].get_ydata()[1]) # 아래쪽 수염의 max, min
print(box['whiskers'][1].get_ydata()[1]) # 위쪽 수염의 min, max
```

* [4-4] 바이올린 플롯(Violin plot) : plt.violinplot(데이터, vert = False)  
    - Boxplot과 비슷하지만, 분포 함께 확인할 수 있다.  
    - plt.violinplot 가로 : vert = False  
    - plt.violinplot 세로 : vert = True
    - 평균선 그려주기 : showmeans = True  

### - 평균값을 텍스트로 그려주기  
plt.text()

```python
# 0) 라이브러리 로딩

import matplotlib.pyplot as plt
import numpy as np

# ----------------------------------------------
# 1) 바이올린 플롯 그리기

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]
age_m = np.mean(age)

plt.violinplot(age, vert = False, showmeans=True)
plt.text( age_m - 1, 0.85 , age_m) # 평균값 텍스트로 그리기
plt.show()
```

```python
# 2) 4가지 플롯을 한꺼번에 그려 봅시다.

age = [19,20,23,46,21,25,26,25,28,31,37,24,28,34,38,33,32,29,27,24]

plt.subplot(2,2,1)
plt.hist(age, bins = 8, edgecolor = 'grey')

plt.subplot(2,2,2)
plt.boxplot(age, vert = False)

plt.subplot(2,2,3)
pd.Series(age).plot(kind = 'kde')

plt.subplot(2,2,4)
plt.violinplot(age, vert = False, showmeans=True)

plt.show()
```

```python
# 3-1) 연습문제를 풀어 봅시다.

import pandas as pd
data = pd.read_csv("https://bit.ly/3HaMAtZ", usecols = ['Age','Fare'])
# Age 컬럼에 대해 

# 4가지 플롯을 한꺼번에 그려 봅시다.
plt.subplot(2, 2, 1)
plt.hist(data['Age'], bins = 50, edgecolor = 'gray')

plt.subplot(2, 2, 2)
plt.boxplot(data['Age'], vert = False)

plt.subplot(2, 2, 3)
pd.Series(data['Age']).plot(kind = 'kde')

plt.subplot(2, 2, 4)
plt.violinplot(data['Age'], vert = False)
plt.text(50, 0.9, round(data['Age'].mean(), 2)) # 평균값 추가

plt.show()
```

```python
# 3-2) 연습문제를 풀어 봅시다.

import pandas as pd
data = pd.read_csv("https://bit.ly/3HaMAtZ", usecols = ['Age','Fare'])
# Fare 컬럼에 대해 

# 4가지 플롯을 한꺼번에 그려 봅시다.
plt.subplot(2, 2, 1)
plt.hist(data['Fare'], bins = 50, edgecolor = 'gray')

plt.subplot(2, 2, 2)
plt.boxplot(data['Fare'], vert = False)

plt.subplot(2, 2, 3)
pd.Series(data['Fare']).plot(kind = 'kde')

plt.subplot(2, 2, 4)
plt.violinplot(data['Fare'], vert = False)
plt.text(50, 0.9, round(data['Fare'].mean(), 2)) # 평균값 추가

plt.show()
```

```python
# 4) 시계열 데이터 시각화
import pandas as pd

air = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/air2.csv ')
air['Date'] = pd.to_datetime(air['Date']) # 날짜 형식으로 변환

plt.plot('Date', 'Ozone', 'g-', data = air, label = 'Ozone')
plt.plot('Date', 'Temp', 'r-', data = air, label = 'Temp')

plt.xlabel('Date')
plt.legend()
plt.show()
```
