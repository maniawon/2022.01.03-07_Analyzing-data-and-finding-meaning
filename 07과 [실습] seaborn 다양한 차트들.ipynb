## 2. seaborn 다양한 차트들

### 1) 기본 차트들

① iris의 Petal.Length 에 대해서, Species 별로 비교하는 차트를 그려봅시다.
    * histplot()
    * kdeplot()
    * boxplot()

* histogram : sns.histplot

```python
sns.histplot(x='Petal.Length' , data = iris, hue = 'Species',
            bins = 16)
plt.show()
```

* densityplot : sns.kdeplot

```python
sns.kdeplot(x='Petal.Length', data = iris, hue = 'Species')
plt.show()
```

* sns.boxplot

```python
sns.boxplot(x='Petal.Length', y = 'Species', data = iris)
plt.show()
```

```python
sns.boxplot(data = iris, y = 'Petal.Length', x = 'Species')
plt.show()
```

### 2) distplot : histogram + density plot

① titanic['Fare'] 에 대해서 distplot을 그려 봅시다.
* bin의 개수를 10 ~ 50 사이에서 조절해 봅시다.
* 히스토그램을 꾸며 봅시다. hist_kws = {'edgecolor':'grey'}


```python
sns.distplot(titanic['Fare'], bins = 40, 
             hist_kws = {'edgecolor':'grey'})
plt.show()
```

② boston['medv'] 에 대해서 distplot을 그려 봅시다.

* bin의 개수를 10 ~ 50 사이에서 조절해 봅시다.
* 히스토그램을 꾸며 봅시다. hist_kws = {'edgecolor':'grey'}

```python
sns.distplot(boston['medv'], bins = 40, 
             hist_kws = {'edgecolor':'grey'})
plt.show()
```

###  3) jointplot : scatter + histogram(혹은 density plot)

① titanin['Age']와 titanic['Fare']에 대해 join plot을 그려봅시다.

```python
sns.jointplot(titanic['Age'], titanic['Fare']) # ← 이렇게 쓰고 hue 옵셥 쓰면 에러 뜸
plt.show()
```

```python
sns.jointplot(x='Age', y='Fare', data = titanic) # ← 이렇게 쓰고 hue 써야 잘나옴
plt.show()
```

② titanin['Age']와 titanic['Fare']에 대해 joint plot을 그려봅시다. 단, 생존여부(Survived) 별로 구분.

```python
# hue 옵션으로 범주 차원을 추가해서 볼 수 있다
sns.jointplot(x='Age', y='Fare', data = titanic, hue = 'Survived')
plt.show()
```

### 7) heatmap : 두 범주 집계 시각화

stock 데이터 : 코스피 지수

```python
stock.info()
```

① [groupby] stock 데이터를 이용하여 year별, month별 평균 주가(close)를 집계하시오.

```python
t1 = stock.groupby(by=['Year', 'Month'], as_index = True)['Close'].mean()
t2 = stock.groupby(by=['Year', 'Month'], as_index = False)['Close'].mean()

print(type(t1), type(t2))
print(t1)
print(t2)
```

```python
# 집계
temp1 = stock.groupby(by=['Year', 'Month'], as_index = False)['Close'].mean()
temp1
```

② [pivot]  'Year','Month', 'Close' 으로 피봇시키시오.

```python
# pivot : 집계.pivot('컬럼1', '컬럼2', '컬럼3', ...)
temp2 = temp1.pivot('Year', 'Month', 'Close')
temp2
```

③ heatmap을 그리시오.

```python
# 그래프 크기 조절 : plt.figure(figsize = (가로, 세로))
plt.figure(figsize = (15, 5)) # 기본사이즈 : 6.4, 4.4

# annot = True : heatmap위에 값 표시
# fmt = '.2f' : 소수점2번째자리까지
# fmt = '.0f' : 정수로 표시
sns.heatmap(temp2, annot = True, fmt = '.0f') 
plt.show()
```
